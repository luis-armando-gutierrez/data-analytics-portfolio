```python



df = pd.read_csv('/datasets/Churn.csv')

# 1) descartar columnas de identificaci√≥n
df = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])

# 2) imputar Tenure (a√±os) y convertir a entero
imp = SimpleImputer(strategy='most_frequent')  # o 'median'
df['Tenure'] = imp.fit_transform(df[['Tenure']]).ravel().astype(int)

# 3) opcional: castear binarios a bool
for col in ['HasCrCard', 'IsActiveMember', 'Exited']:
    df[col] = df[col].astype(bool)

# 4) opcional: marcar categ√≥ricas como category (para EDA/memoria)
df['Geography'] = df['Geography'].astype('category')
df['Gender'] = df['Gender'].astype('category')

print(df.dtypes)
print(df.isnull().sum())
```

### Upsampling

```python

df_ohe = pd.get_dummies(df, drop_first=True)

y = df_ohe['Exited']
X = df_ohe.drop('Exited', axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, random_state=12345, stratify=y
)
```

```python
# Funci√≥n de upsampling (replicar clase 1 hasta igualar a la 0)
def upsample_minority(X, y, random_state=12345):
    data = X.copy()
    data['Exited'] = y.values

    maj = data[data['Exited'] == 0]
    mino = data[data['Exited'] == 1]

    # replicar clase 1 hasta el tama√±o de la clase 0
    reps = int(len(maj) / len(mino))
    rem  = len(maj) - reps * len(mino)
    mino_rep = pd.concat([mino]*reps + [mino.sample(rem, random_state=random_state)], axis=0)

    up = pd.concat([maj, mino_rep], axis=0)
    up = shuffle(up, random_state=random_state).reset_index(drop=True)

    X_up = up.drop('Exited', axis=1)
    y_up = up['Exited']
    return X_up, y_up

X_train_up, y_train_up = upsample_minority(X_train, y_train)
print("Tama√±os tras upsampling:", X_train_up.shape, y_train_up.value_counts().to_dict())
```

```python
#Entrenar y evaluar
# Modelo con datos upsampled
lr_up = LogisticRegression(max_iter=1000, random_state=12345)
lr_up.fit(X_train_up, y_train_up)

proba_valid = lr_up.predict_proba(X_valid)[:, 1]
pred_valid  = (proba_valid >= 0.5).astype(int)

print("Logistic Regression con upsampling")
print("F1  :", round(f1_score(y_valid, pred_valid), 3))
print("AUC :", round(roc_auc_score(y_valid, proba_valid), 3))

```

El F1 ya subi√≥ mucho m√°s frente al baseline (0.069). el upsampling ayud√≥ a detectar m√°s casos de Exited=1.

El AUC (0.732) indica que el modelo s√≠ separa razonablemente a ambas clases a nivel de probabilidades (independiente del umbral).

```python
# Afianr umbral para maximizar f1

best_t, best_f1 = 0.5, -1
for t in np.arange(0.1, 0.9, 0.01):
    f1 = f1_score(y_valid, (proba_valid >= t).astype(int))
    if f1 > best_f1:
        best_f1, best_t = f1, t

print("Mejor umbral:", round(best_t,2), " | F1:", round(best_f1,3))
```

### Class weight Balanceado

```python
df_ohe = pd.get_dummies(df, drop_first=True)

y = df_ohe['Exited']
X = df_ohe.drop('Exited', axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, random_state=12345, stratify=y
)

# util para buscar el mejor umbral por F1
def best_threshold(proba, y_true, lo=0.1, hi=0.9, step=0.01):
    best_t, best_f1 = 0.5, -1
    for t in np.arange(lo, hi, step):
        f1 = f1_score(y_true, (proba >= t).astype(int))
        if f1 > best_f1:
            best_f1, best_t = f1, t
    return best_t, best_f1


#  1) Logistic Regression con class_weight='balanced' 
lr_bal = LogisticRegression(max_iter=1000, random_state=12345, class_weight='balanced')
lr_bal.fit(X_train, y_train)

proba_lr = lr_bal.predict_proba(X_valid)[:, 1]
pred_lr_050 = (proba_lr >= 0.50).astype(int)

print("LogisticRegression (class_weight='balanced')")
print("F1  @0.50:", round(f1_score(y_valid, pred_lr_050), 3))
print("AUC     :", round(roc_auc_score(y_valid, proba_lr), 3))

t_lr, f1_lr = best_threshold(proba_lr, y_valid, lo=0.2, hi=0.8, step=0.01)
print("Mejor umbral LR:", round(t_lr, 2), "| F1:", round(f1_lr, 3))
print()


#  2) RandomForest con class_weight='balanced' 
rf_bal = RandomForestClassifier(
    n_estimators=300, max_depth=10, random_state=12345, n_jobs=-1,
    class_weight='balanced'
)
rf_bal.fit(X_train, y_train)

proba_rf = rf_bal.predict_proba(X_valid)[:, 1]
pred_rf_050 = (proba_rf >= 0.50).astype(int)

print("RandomForest (class_weight='balanced')")
print("F1  @0.50:", round(f1_score(y_valid, pred_rf_050), 3))
print("AUC     :", round(roc_auc_score(y_valid, proba_rf), 3))

t_rf, f1_rf = best_threshold(proba_rf, y_valid, lo=0.2, hi=0.8, step=0.01)
print("Mejor umbral RF:", round(t_rf, 2), "| F1:", round(f1_rf, 3))
```

LogisticRegression (balanced): F1‚âà0.46, AUC‚âà0.72  mejora vs baseline, pero se queda corto.

RandomForest (balanced): F1‚âà0.64 @0.48, AUC‚âà0.87  supera holgadamente la meta (‚â•0.59). Adem√°s, el ajuste de umbral (‚âà0.48) mejora F1 respecto a 0.50.

Conclusi√≥n: el mejor enfoque es RandomForest con class_weight='balanced' y umbral ‚âà 0.48.

```python
df_ohe = pd.get_dummies(df, drop_first=True)

y = df_ohe['Exited']
X = df_ohe.drop('Exited', axis=1)

# train/valid/test = 60/20/20
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.20, random_state=12345, stratify=y
)
X_train, X_valid, y_train, y_valid = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=12345, stratify=y_temp
)

# reentrenar en train+valid con el mejor modelo
X_trfinal = pd.concat([X_train, X_valid], axis=0)
y_trfinal = pd.concat([y_train, y_valid], axis=0)

best_rf = RandomForestClassifier(
    n_estimators=300, max_depth=10, random_state=12345,
    n_jobs=-1, class_weight='balanced'
)
best_rf.fit(X_trfinal, y_trfinal)

# evaluar en test con el umbral encontrado en valid (‚âà 0.48)
proba_test = best_rf.predict_proba(X_test)[:, 1]
thr = 0.48
pred_test = (proba_test >= thr).astype(int)

print("TEST F1 :", round(f1_score(y_test, pred_test), 3))
print("TEST AUC:", round(roc_auc_score(y_test, proba_test), 3))
print(classification_report(y_test, pred_test, digits=3))
```
